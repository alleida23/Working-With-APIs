```python
url = 'https://newsapi.org/v2/everything?domains=wsj.com&apiKey=f0891beff53347f297c4ce12326063d1'

response = requests.get(url)

if response.status_code == 200:
    data = response.json()['articles']
    print('Successful request with status code',response.status_code)
else:
    print('Request failed with status code', response.status_code)
```

#### Polygon.io

```python
import getpass
API_key = getpass.getpass()
```

```python
import requests

API_key
url = 'https://api.polygon.io/v2/reference/news'

params = {
    'apiKey': API_key,
    'assetClass': 'indices',
    'locale': 'us'
}

# Send a GET request to the API with the specified parameters
response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    articles = data['results'] # Extracting articles from the response

    # Printing titles and url of the articles
    for article in articles:
        print(f"Title: {article['title']}")
        print(f"URL: {article['article_url']}")
        print('-' * 40) # adding ----- separator
else:
    print(f"Error: {response.status_code} - {response.text}")
```

```python
#articles
```

```python
# Json_normalize articles
polygon_flattened_data = json_normalize(articles)
polygon_flattened_data.head(5)
```

```python
# Dataframe
polygon_data = pd.DataFrame(polygon_flattened_data)
polygon_data.head(5)
```

```python
# Dropping columns to match the ones from wall_data
polygon_data = polygon_data.drop(['id', 'tickers','amp_url', 'image_url',
                                  'keywords', 'publisher.homepage_url',
                                  'publisher.logo_url', 'publisher.favicon_url'], axis=1)
```

```python
# Renaming columns to match the ones from wall_data
polygon_data = polygon_data.rename(columns={'published_utc':'published_at',
                                            'publisher.name': 'source',
                                           'article_url':'url'})

polygon_data.columns
```

```python
# Rearranging columns to match the ones from polygon_data
polygon_data = polygon_data[['author', 'title', 'description', 'url', 'published_at', 'source']]

display(polygon_data.columns)
display(wall_data.columns)
```

#### Advice Slip JSON API

```python
import random
import time

url = 'https://api.adviceslip.com/advice'

# Getting and collecting random advices
advices = []

# This response gets only 1 advice each time,
# and I'll need one per row of news_data dataframe.

for i in range(len(news_data)): 
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()['slip']['advice']
        advices.append(data)
        print(f'Successful request {i+1} with status code {response.status_code}')
    else:
        print(f'Request {i+1} failed with status code {response.status_code}')
    
    # Generating random sleep_tome to avoid any block
    sleep_time = random.uniform(0.5, 2.0) # Generate a random sleep time between 0.5 and 2.0 seconds
    time.sleep(sleep_time) # Sleep for the generated time
```
